Ejercicio 1

```python
def f(x): #definimos la funcion
    return 2*x - np.log(x) - 4

def Df(x): #definimos la derivada de la funcion
    return 2 - 1/x
    
i = 1
x0 = 1
for iteracion in range (1, 4): #recorremos f(x) con un bucle for 
    x1 = x0 - f(x0) / Df(x0) #formula de Newton
    x0 = x1 #igualamos x0 a x1 para que se reemplace el valor 
    print("Iteracion",i,", raiz aproximada: ", x0)
    i = i + 1

    m = np.linspace(1, 10, 30) #creamos la grafica de iteraciones
    y = f(m)
    my.plot(m,y)
    
print("Grafica de iteraciones") 
```

Ejercicio 2

```python
def f(x): #definimos la funcion
    return x**3 - 2*x - 5

def Df(x): #definimos la derivada de la funcion
    return 3*x**2 - 2

x0 = 0
i = 1

for iteracion in range (1, 4): #recorremos f(x) con un bucle for 
    x1 = x0 - f(x0) / Df(x0) #formula de Newton
    x0 = x1 #igualamos x0 a x1 para que se reemplace el valor 
    print("Iteracion",i,", raiz aproximada: ", x0)
    i = i + 1
    
    m = np.linspace(1, 10, 30) #creamos la grafica de iteraciones
    y = f(m)
    my.plot(m,y)
    
print("Grafica de iteraciones")
```

Ejercicio 3

```python
def f(x): #definimos la funcion
    return 2**x - 3*x

def Df(x):  #definimos la derivada de la funcion
    return 2**x*np.log(2) - 3

x0 = 0
i = 1
for iteracion in range (1, 4):
    x1 = x0 - f(x0) / Df(x0) #formula de Newton
    x0 = x1 #igualamos x0 a x1 para que se reemplace el valor 
    print("Iteracion",i,", raiz aproximada: ", x0)
    i = i + 1
    
    m = np.linspace(-1, 10, 40) #creamos la grafica de iteraciones
    y = f(m)
    my.plot(m,y)
    
print("Grafica de iteraciones")
```

Grafica de la funcion
